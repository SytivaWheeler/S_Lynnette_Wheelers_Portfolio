Reading Datasets

```python
import pandas as pd
confirmed=pd.read_csv("../../../../data/stage_1/covid_confirmed_usafacts.csv")
population=pd.read_csv("../../../../data/stage_1/covid_county_population_usafacts.csv")
deaths=pd.read_csv("../../../../data/stage_1/covid_deaths_usafacts.csv")


confirmed_cases=pd.DataFrame(confirmed)
population_county=pd.DataFrame(population)
death_count=pd.DataFrame(deaths)
```

Merging all datasets into a super dataset

```python
merge_1=pd.merge(confirmed_cases,death_count,on=['countyFIPS','County Name','State','StateFIPS'])
merge_1.dtypes
merge_2=pd.merge(population_county,merge_1,on=['countyFIPS','State'])
```

Saving the superset dataframe to .csv file

```python
merge_2.to_csv("../../../../data/stage_1/superset.csv",index=False)
```
