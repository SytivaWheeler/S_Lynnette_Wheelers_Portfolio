Reading of Enrichment-ACS Demographic and superset of Covid-19 datasets

```python
import pandas as pd
enr=pd.read_csv("../../../../data/stage_1/ACSDP5Y2019.DP05_data_with_overlays_2021-09-09T150852.csv")
acs_data=pd.DataFrame(enr)
covid=pd.read_csv("../../../../data/stage_1/superset.csv")


acs_data
covid
```

Splitting the first two by taking them into new datasets

```python
acs_data=acs_data.drop([0])
acs_data['GEO_ID']=acs_data['GEO_ID'].str.slice(start=9)
```

Merging data back into a single dataset making necessary changes like droping,renaming,changing datatype of some columns and dropped few unnecessary rows 

```python
acs_data[['County Name','State']]=pd.DataFrame(acs_data.NAME.str.split(',',expand=True))
```

```python

#dropped Puerto Rico related data
acs_data['GEO_ID']=acs_data['GEO_ID'].astype('int64')
acs_data=acs_data[acs_data['GEO_ID']<56042]

```

```python
acs_final_filter=acs_data[['GEO_ID','County Name','State','DP05_0002E','DP05_0002PE','DP05_0003E','DP05_0003PE','DP05_0005E','DP05_0006E','DP05_0007E',
                               'DP05_0008E','DP05_0009E','DP05_0010E','DP05_0011E','DP05_0012E','DP05_0013E','DP05_0014E','DP05_0015E',
                               'DP05_0016E','DP05_0017E','DP05_0018E','DP05_0030E','DP05_0031E','DP05_0037E','DP05_0038E',
                               'DP05_0039E','DP05_0044E','DP05_0052E']]
acs_final_filter
#acs_final=acs_final.rename(columns={'GEO_ID':'CountyFIPS',})
```

Merged enrichment-ACS data and Covid-19 Superset based on CountyFIPS

```python

final_merge=pd.merge(covid,acs_final_filter,how='inner', left_on='countyFIPS', right_on='GEO_ID')
final_merge
```

```python
final_merge=final_merge.rename(columns={'State_x':'State','County Name_x':'County Name','DP05_0002E':'Total_Males','DP05_0002PE':'Male_percent','DP05_0003E':'Total_Females','DP05_0003PE':'Female_percent','DP05_0005E':'under 5','DP05_0006E':'5 to 9','DP05_0007E':'10 to 14',
                               'DP05_0008E':'15 to 19','DP05_0009E':'20 to 24','DP05_0010E':'25 to 34','DP05_0011E':'35 to 44','DP05_0012E':'45 to 54','DP05_0013E':'55 to 59','DP05_0014E':'60 to 64','DP05_0015E':'65 to 74',
                               'DP05_0016E':'75 to 84','DP05_0017E':'above85','DP05_0018E':'Median_Age','DP05_0030E':'Male_above65','DP05_0031E':'Female_above65','DP05_0037E':'White_population','DP05_0038E':'Black_population',
                               'DP05_0039E':'American_Indian_and_Alaska_Native','DP05_0044E':'Asian_population','DP05_0052E':'Native_Hawaiian_and_Other Pacific_Islander'})
```
