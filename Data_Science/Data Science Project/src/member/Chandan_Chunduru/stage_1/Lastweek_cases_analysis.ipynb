Reading Confirmed Cases,Death Cases and Converting to Data Frame

```python
import pandas as pd
confirmed=pd.read_csv("../../../../data/stage_1/covid_confirmed_usafacts.csv")
confirmed_cases=pd.DataFrame(confirmed)
deaths=pd.read_csv("../../../../data/stage_1/covid_deaths_usafacts.csv")
deaths_count=pd.DataFrame(deaths)
```

Selection of cases,deaths for last 7 days in the state of Texas

```python
confirmed_cases=confirmed_cases[confirmed_cases.State=='TX']
confirmed_cases_last_week=confirmed_cases.iloc[:,-7:]
deaths_count=deaths_count[deaths_count.State=='TX']
deaths_last_week=deaths_count.iloc[:,-7:]
```

Adding up of all cases for a date

```python
confirmed_cases_last_week=confirmed_cases_last_week.sum(axis=0)
deaths_last_week=deaths_last_week.sum(axis=0)
```

Plot for the last week data

```python
confirmed_cases_last_week.plot()
```

From the plot it is  observed that cases are increasing over past 7 days.

```python
deaths_last_week.plot()
```
