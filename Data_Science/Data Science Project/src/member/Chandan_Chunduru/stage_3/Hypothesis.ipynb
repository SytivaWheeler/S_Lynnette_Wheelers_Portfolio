```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
%matplotlib inline
from sklearn.linear_model import LinearRegression 
import statsmodels.formula.api as smf
import statsmodels.tools.eval_measures as m
```

```python
Data=pd.read_csv("../../../../data/stage_1/EC.csv")
#Data['Date']=Data['Date'].astype('datetime64[ns]')
```

```python
Data=Data.drop(['GEO_ID','County Name_y','State_y','County Name_y'],axis=1)
```

```python
final=Data[Data['Date']=='2021-08-16']
```

```python
final
```

```python
final=final.groupby(['State']).sum().reset_index()
final
```

```python
stats.ttest_ind(a= final.Black_population,
                b= final.White_population,
                alternative='two-sided') 
```

P value is less than 0.05 so we can reject this null hypothesis

Are people above 85 more prone to covid than people of 25 to 34

```python
stats.ttest_ind(a= final.above85,
                b= final['25 to 34'],
                alternative='two-sided') 
```

P value is less than 0.05 so we can reject  this null hypothesis also

children below 5 years are less effected by covid

```python
stats.ttest_ind(a= final.Cases,
                b= final['under 5'],
                alternative='two-sided') 
```

P value is less than 0.05 so we can reject this null hypothesis also

```python
stats.ttest_ind(a= final.Cases,
                b= final.Black_population,
                equal_var=False) 
```

p value is greater than .05 so we can't reject this hypothesis

```python
stats.ttest_ind(a= final.Cases,
                b= final.White_population,
                equal_var=False) 
```
